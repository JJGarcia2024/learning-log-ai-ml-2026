```python
# Your code here

# Query to find all distinct users who posted on January 1, 2019.
# We need to check BOTH the questions table AND the answers table,
# then combine them with UNION DISTINCT to remove duplicates.
all_users_query = """
                  -- Get user IDs of everyone who ASKED a question on Jan 1, 2019
                  SELECT owner_user_id
                  FROM `bigquery-public-data.stackoverflow.posts_questions`
                  WHERE EXTRACT(DATE FROM creation_date) = '2019-01-01'

                  -- UNION DISTINCT combines both result sets and automatically
                  -- removes any duplicate user IDs (e.g. someone who both asked
                  -- AND answered on the same day will only appear once)
                  UNION DISTINCT

                  -- Get user IDs of everyone who ANSWERED a question on Jan 1, 2019
                  SELECT owner_user_id
                  FROM `bigquery-public-data.stackoverflow.posts_answers`
                  WHERE EXTRACT(DATE FROM creation_date) = '2019-01-01'
                  """

# Set up the query (cancel the query if it would use too much data)
safe_config = bigquery.QueryJobConfig(maximum_bytes_billed=10**10)

# Run the query and convert results to a pandas DataFrame
# (create_bqstorage_client=False bypasses the Storage API permission error)
all_users = client.query(all_users_query, job_config=safe_config).result().to_dataframe(create_bqstorage_client=False)

# Print the total number of distinct users who posted on Jan 1, 2019
print("Number of distinct users who posted on Jan 1, 2019:", len(all_users))

# Preview the first few rows
print(all_users.head())

# Check your answer
q_4.check()
```

```python
# 6. Building a more generally useful service
def expert_finder(topic, client):
    '''
    Returns a DataFrame with the user IDs who have written answers on a given topic.
    '''
    
    # We use an f-string (notice the 'f' before the quotes) 
    # to dynamically inject our topic variable into the SQL query.
    my_query = f"""
               SELECT a.owner_user_id AS user_id, COUNT(1) AS number_of_answers
               FROM `bigquery-public-data.stackoverflow.posts_questions` AS q
               INNER JOIN `bigquery-public-data.stackoverflow.posts_answers` AS a
                   ON q.id = a.parent_id
               WHERE q.tags LIKE '%{topic}%'
               GROUP BY a.owner_user_id
               ORDER BY number_of_answers DESC
               """
               
    # Set up the query
    safe_config = bigquery.QueryJobConfig(maximum_bytes_billed=10**10)
    my_query_job = client.query(my_query, job_config=safe_config)
    
    # API request - run the query, and return a pandas DataFrame
    results = my_query_job.to_dataframe(create_bqstorage_client=False)
    
    return results

# Let's test it out by finding experts on 'python' instead of 'bigquery'!
python_experts = expert_finder('python', client)
print(python_experts.head())
```
