```python
# 2) How many unique types of locations are there in the crime table?

query = """
        SELECT COUNT(DISTINCT location_description) AS unique_locations
        FROM `bigquery-public-data.chicago_crime.crime`
        """

query_job = client.query(query)
result = query_job.to_dataframe()
unique_locations = result['unique_locations'][0]
print(f"Unique location types: {unique_locations}")

# Check your answer
q_2.check()
```

```python
# 3) Write a query that shows the number of crimes per year, ordered from oldest to most recent.

query = """
        SELECT year, COUNT(1) AS num_crimes
        FROM `bigquery-public-data.chicago_crime.crime`
        GROUP BY year
        ORDER BY year
        """

query_job = client.query(query)
crimes_by_year = query_job.to_dataframe()
print(crimes_by_year)

# Check your answer
q_3.check()
```

```python
# Bonus: Which primary crime types appear most frequently?

query = """
        SELECT primary_type, COUNT(1) AS crime_count
        FROM `bigquery-public-data.chicago_crime.crime`
        GROUP BY primary_type
        ORDER BY crime_count DESC
        LIMIT 10
        """

query_job = client.query(query)
top_crimes = query_job.to_dataframe()
print(top_crimes)
```
